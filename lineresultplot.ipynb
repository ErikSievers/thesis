```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pathlib
from copy import copy
from matplotlib import cm, colors
import cv2

bins = [5,10,20,40,80]

baseline05 = 0.966
baseline025 = 0.95
baseline01 = 0.819

nameset = set()

####################
# Start
####################

baseline = baseline01
dtresult = pd.read_csv('./Results/DT-0.1-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
knnresult = pd.read_csv('./Results/KNN-0.1-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)

# Extract "good" settings?
dtgoodsettings = dtresult[dtresult['test-auc'] > baseline]
knngoodsettings = knnresult[knnresult['test-auc'] > baseline]

goodsettingsmap = {}
for index, row in dtgoodsettings.iterrows():
    key = ('DT', row['type'], row['windowSize'], row['neighbourhood'])
    if key not in goodsettingsmap:
        goodsettingsmap[key] = [
            dtresult[(dtresult['buckets'] == b) & (dtresult['type'] == row['type']) & (dtresult['windowSize'] == row['windowSize']) & (dtresult['neighbourhood'] == row['neighbourhood'])]['test-auc'].max()
            for b in bins
        ]
for index, row in knngoodsettings.iterrows():
    key = ('KNN', row['type'], row['windowSize'], row['neighbourhood'])
    if key not in goodsettingsmap:
        goodsettingsmap[key] = [
            knnresult[(knnresult['buckets'] == b) & (knnresult['type'] == row['type']) & (knnresult['windowSize'] == row['windowSize']) & (knnresult['neighbourhood'] == row['neighbourhood'])]['test-auc'].max()
            for b in bins
        ]
plt.plot(bins, [baseline]*len(bins), label='Baseline classifier', color='black', linestyle='dashed')
for (classifier, type, z, xy), values in goodsettingsmap.items():
    nameset.add((classifier, type, z, xy))
    plt.plot(bins, values, label=f'{classifier} {type} Z={z} XY={xy}')
plt.xscale('log')
plt.xlabel("Number of bins")
plt.ylabel("ROC-AUC")
plt.xticks(bins, bins)
plt.legend()
plt.figure()


####################
# Done with 0.1
####################


baseline = baseline025
dtresult = pd.read_csv('./Results/DT-0.25-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
knnresult = pd.read_csv('./Results/KNN-0.25-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)

# Extract "good" settings?
dtgoodsettings = dtresult[dtresult['test-auc'] > baseline]
knngoodsettings = knnresult[knnresult['test-auc'] > baseline]

goodsettingsmap = {}
for index, row in dtgoodsettings.iterrows():
    key = ('DT', row['type'], row['windowSize'], row['neighbourhood'])
    if key not in goodsettingsmap:
        goodsettingsmap[key] = [
            dtresult[(dtresult['buckets'] == b) & (dtresult['type'] == row['type']) & (dtresult['windowSize'] == row['windowSize']) & (dtresult['neighbourhood'] == row['neighbourhood'])]['test-auc'].max()
            for b in bins
        ]
for index, row in knngoodsettings.iterrows():
    key = ('KNN', row['type'], row['windowSize'], row['neighbourhood'])
    if key not in goodsettingsmap:
        goodsettingsmap[key] = [
            knnresult[(knnresult['buckets'] == b) & (knnresult['type'] == row['type']) & (knnresult['windowSize'] == row['windowSize']) & (knnresult['neighbourhood'] == row['neighbourhood'])]['test-auc'].max()
            for b in bins
        ]
plt.plot(bins, [baseline]*len(bins), label='Baseline classifier', color='black', linestyle='dashed')
for (classifier, type, z, xy), values in goodsettingsmap.items():
    nameset.add((classifier, type, z, xy))
    plt.plot(bins, values, label=f'{classifier} {type} Z={z} XY={xy}')
plt.xscale('log')
plt.xlabel("Number of bins")
plt.ylabel("ROC-AUC")
plt.xticks(bins, bins)
plt.legend()
plt.figure()

####################
# Done with 0.25
####################

baseline = baseline05
dtresult = pd.read_csv('./Results/DT-0.5-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
knnresult = pd.read_csv('./Results/KNN-0.5-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)

# Extract "good" settings?
dtgoodsettings = dtresult[dtresult['test-auc'] > baseline]
knngoodsettings = knnresult[knnresult['test-auc'] > baseline]

goodsettingsmap = {}
for index, row in dtgoodsettings.iterrows():
    key = ('DT', row['type'], row['windowSize'], row['neighbourhood'])
    if key not in goodsettingsmap:
        goodsettingsmap[key] = [
            dtresult[(dtresult['buckets'] == b) & (dtresult['type'] == row['type']) & (dtresult['windowSize'] == row['windowSize']) & (dtresult['neighbourhood'] == row['neighbourhood'])]['test-auc'].max()
            for b in bins
        ]
for index, row in knngoodsettings.iterrows():
    key = ('KNN', row['type'], row['windowSize'], row['neighbourhood'])
    if key not in goodsettingsmap:
        goodsettingsmap[key] = [
            knnresult[(knnresult['buckets'] == b) & (knnresult['type'] == row['type']) & (knnresult['windowSize'] == row['windowSize']) & (knnresult['neighbourhood'] == row['neighbourhood'])]['test-auc'].max()
            for b in bins
        ]
plt.plot(bins, [baseline]*len(bins), label='Baseline classifier', color='black', linestyle='dashed')
for (classifier, type, z, xy), values in goodsettingsmap.items():
    nameset.add((classifier, type, z, xy))
    plt.plot(bins, values, label=f'{classifier} {type} Z={z} XY={xy}')
plt.xscale('log')
plt.xlabel("Number of bins")
plt.ylabel("ROC-AUC")
plt.xticks(bins, bins)
plt.legend()
plt.figure()

####################
# Done with 0.5
####################

# Three plots, one per threshold
```
