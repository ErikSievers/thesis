```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pathlib
from copy import copy
from matplotlib import cm, colors
import cv2

bins = [5,10,20,40,80]

baseline05 = 0.966
baseline025 = 0.95
baseline01 = 0.819

baseline = baseline05
dtresult = pd.read_csv('./Results/DT-0.5-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
knnresult = pd.read_csv('./Results/KNN-0.5-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
dtresult.insert(1, 'Classifier', 'DT')
knnresult.insert(1, 'Classifier', 'KNN')
masterdf = pd.concat([dtresult, knnresult])
masterdf.insert(1, "Threshold", 0.5)

####################
# Done with 0.5
####################

baseline = baseline025
dtresult = pd.read_csv('./Results/DT-0.25-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
knnresult = pd.read_csv('./Results/KNN-0.25-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
dtresult.insert(1, 'Classifier', 'DT')
knnresult.insert(1, 'Classifier', 'KNN')
dtresult.insert(1, 'Threshold', 0.25)
knnresult.insert(1, 'Threshold', 0.25)
masterdf = pd.concat([masterdf, knnresult, dtresult])

####################
# Done with 0.25
####################

baseline = baseline01
dtresult = pd.read_csv('./Results/DT-0.1-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
knnresult = pd.read_csv('./Results/KNN-0.1-os-1-usr-1-pm-1.csv',
    names=['type','sharpening','windowSize','neighbourhood','buckets','histnormalise','k-nearest','cv-auc','test-auc','test-precision','test-recall']
    , header=1)
dtresult.insert(1, 'Classifier', 'DT')
knnresult.insert(1, 'Classifier', 'KNN')
dtresult.insert(1, 'Threshold', 0.1)
knnresult.insert(1, 'Threshold', 0.1)
masterdf = pd.concat([masterdf, knnresult, dtresult])
masterdf = masterdf.sort_values(by=['cv-auc'], ascending=False)

####################

baselines = {0.1: 0.819, 0.25: 0.95, 0.5: 0.966}

print("   --- CV based on Z ---")
for threshold in [0.1, 0.25, 0.5]:
    for classifier in ['DT', 'KNN']:
        for Z in [1,3,5,7]:
            res = masterdf[(masterdf['Classifier'] == classifier) & (masterdf['Threshold'] == threshold) & (masterdf['windowSize'] == Z)]['test-auc'].iloc[0]
            print(f'{threshold} {classifier} {Z} {res} {res - baselines[threshold]}')

print("")
print("   --- CV based on XY ---")

for threshold in [0.1, 0.25, 0.5]:
    for classifier in ['DT', 'KNN']:
        for XY in [3,5,7]:
            res = masterdf[(masterdf['Classifier'] == classifier) & (masterdf['Threshold'] == threshold) & (masterdf['neighbourhood'] == XY)]['test-auc'].iloc[0]
            print(f'{threshold} {classifier} {XY} {res} {res - baselines[threshold]}')

print("")
print("   --- CV based on bins ---")

for threshold in [0.1, 0.25, 0.5]:
    for classifier in ['DT', 'KNN']:
        for bins in [5,10,20,40,80]:
            res = masterdf[(masterdf['Classifier'] == classifier) & (masterdf['Threshold'] == threshold) & (masterdf['buckets'] == bins)]['test-auc'].iloc[0]
            print(f'{threshold} {classifier} {bins} {res} {res - baselines[threshold]}')

print("")
print("   --- CV based on XYZ ---")

for threshold in [0.1, 0.25, 0.5]:
    for classifier in ['DT', 'KNN']:
        for Z in [1,3,5,7]:
            for XY in [3,5,7]:
                res = masterdf[(masterdf['Classifier'] == classifier) & (masterdf['Threshold'] == threshold) & (masterdf['windowSize'] == Z) & (masterdf['neighbourhood'] == XY)].iloc[0]
                print(f'{threshold} {classifier} {Z} {XY} {res} {res - baselines[threshold]}')
```

```python
print("threshold classifier method Z XY bins CV test")
columns = ['Threshold', 'Classifier', 'type', 'Z', 'XY', 'bins', 'cv-auc', 'test-auc']
results = pd.DataFrame(columns=columns)
for threshold in [0.1, 0.25, 0.5]:
    for classifier in ['DT', 'KNN']:
        for method in ['moran', 'scatter', 'spatstat']:
            for Z in [1,3,5,7]:
                for XY in [3,5,7]:
                    res = masterdf[(masterdf['Classifier'] == classifier) & (masterdf['Threshold'] == threshold)
                        & (masterdf['type'] == method)
                        & (masterdf['windowSize'] == Z) & (masterdf['neighbourhood'] == XY)].iloc[0]
                    test = res['test-auc']
                    cv = res['cv-auc']
                    bins = res['buckets']
                    nextRow =  pd.DataFrame([[threshold, classifier, method, Z, XY, bins, cv, test]], columns=columns)
                    results = pd.concat([results, nextRow])

results = results.sort_values(by=['test-auc'], ascending=False)
for threshold in [0.1, 0.25, 0.5]:
    for classifier in ['DT', 'KNN']:
        for method in ['moran', 'scatter', 'spatstat']:
            res = results[(results['Classifier'] == classifier) & (results['Threshold'] == threshold)
                & (results['type'] == method)].iloc[0]
            Z = res['Z']
            XY = res['XY']
            test = res['test-auc']
            cv = res['cv-auc']
            bins = res['bins']
            print(f'{threshold} {classifier} {method} {Z} {XY} {bins} {cv} {test}')
            
```
